```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
study_hours = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
marks = np.array([20,30,40,50,60,70,80,90])

# Model create
model = LinearRegression()

# Train
model.fit(study_hours, marks)

# Prediction line
predicted_marks = model.predict(study_hours)

# Graph
plt.scatter(study_hours, marks)
plt.plot(study_hours, predicted_marks, color='red')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Student Mark Prediction")
plt.show()

# Test prediction
hours = float(input("Enter study hours: "))
prediction = model.predict([[hours]])
print("Predicted Marks:", prediction[0])
```

# Student Mark Prediction using Machine Learning

This project predicts student marks based on study hours using Linear Regression algorithm.
